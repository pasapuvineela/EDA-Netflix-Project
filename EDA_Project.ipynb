```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("netflix_titles.csv")
df.head()
df.shape
df.columns
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["country"] = df["country"].fillna("Unknown")
```

```python
df["director"] = df["director"].fillna("Not Available")
```

```python
df["cast"] = df["cast"].fillna("Not Available")
```

```python
df.isnull().sum()
```

```python
df['type'].value_counts()
```

```python
df['country'].value_counts().head(10)
```

```python
df['rating'].value_counts()
```

```python
df['release_year'].value_counts().sort_index()
```

```python
df['director'].value_counts().head(10)
```

```python
df['type'].value_counts().plot(kind='bar')

plt.title("Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(df['release_year'])

plt.title("Release Year Distribution")
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.show()
```

```python
df['type'].value_counts().plot(
kind='pie',
autopct='%1.1f%%'
)

plt.title("Movies vs TV Shows")
plt.ylabel("")
plt.show()
```

```python
plt.scatter(df['release_year'], df.index)

plt.title("Release Year Scatter")
plt.xlabel("Release Year")
plt.ylabel("Index")
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot()

plt.title("Content Released Per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
sns.heatmap(df.isnull(), cbar=False)

plt.title("Missing Values Heatmap")
plt.show()
```

**Insights Report**

1. Netflix has significantly more Movies than TV Shows, indicating movies form the largest portion of its content library.

2. The United States has the highest number of Netflix titles, making it the leading content-producing country.

3. The number of titles released on Netflix increased rapidly after 2015, showing the platform's expansion.

4. TV-MA is one of the most common ratings, suggesting Netflix primarily targets mature audiences.

5. Missing values were mainly found in the director and cast columns before data cleaning, indicating incomplete metadata.

**Business Recommendations**
1. Increase investment in TV Shows if audience demand continues to grow.

2. Expand content production in countries with fewer Netflix titles to attract global audiences.

3. Focus on producing content in popular genres to improve viewer engagement.

4. Improve data quality by reducing missing information such as director and cast details.

5. Continue adding new content every year to maintain subscriber interest and platform growth.

**Surprising Finding**
The most surprising finding was the rapid increase in Netflix content after 2015. I also observed that Movies greatly outnumber TV Shows. Additionally, the United States contributes the largest share of Netflix titles. This shows how quickly Netflix expanded its content library over recent years.
